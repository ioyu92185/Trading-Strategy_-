```python
import pandas as pd
import csv
import jieba
import re
```

```python
#讀入台大情緒分數檔案
score_table= pd.read_csv('score.csv')
```

```python
#將情緒分數建立成字典
sentment_dict = dict(zip(list(score_table.word),list(score_table.poscore)))
```

```python
#用該字典進行結巴斷詞
for w in sentment_dict.keys():
    jieba.suggest_freq(w,True)
```

```python
#去除停用字
stop_words = [re.findall(r'\S+',w)[0] for w in open('stopWords.txt',encoding='utf8').readlines() if len(re.findall(r'\S+',w))>0]

#將結巴斷詞、去除停用字的步驟寫成函數
def sent2word(sentence,stop_words=stop_words):
    words = jieba.cut(sentence, HMM=False)
    words = [w for w in words if w not in stop_words]
    return words
```

```python
#經由 sent2word 斷詞後，根據斷詞結果對應其分數，累加為該篇情緒分數
def get_sentment(sent):
    tokens = sent2word(sent)
    poscore = 0
    countword = 0
    for w in tokens:
        
        if w in sentment_dict.keys():
            poscore += sentment_dict[w]
            countword += 1
    if countword != 0:
        return poscore
    else:
        return 0
```

```python
#處理文本，去除多餘空行
file1 = open('匯率新聞.csv', 'r', encoding='utf-8') # 要去掉空行的文件 
file2 = open('匯率新聞_1.csv', 'w', encoding='utf-8') # 生成没有空行的文件
try:
    for line in file1.readlines():
        if line == '\n':
            line = line.strip("\n")
        file2.write(line)
finally:
    file1.close()
    file2.close()
```

```python
#讀入文本為 dataframe 格式
news= pd.read_csv('匯率新聞_1.csv')
news
```

```python
#使用上面的函數，計算每篇標題、內文、匯銀主管情緒分數
news['caltext'] = news['text'].apply(get_sentment)
news['caltitle'] = news['title'].apply(get_sentment)
news['calsuper'] = news['匯銀主管'].apply(get_sentment)
news
```

```python
#將上面資料去除文本，並依時間排序
import datetime
n=news
n=n.drop(labels=["title","text","匯銀主管"], axis="columns")
n['date'] = pd.to_datetime(n['date'])
n.sort_values("date",inplace=True)
n
```

```python
#將同一天的新聞情緒分數累加，並將 date 改為 Date（方便等下合併） 
n=n.groupby('date').sum()
n.to_csv("sign.csv",sep=',')
trend= pd.read_csv('sign.csv')
trend = trend.rename(columns={'date': 'Date'})
trend 
```

```python
#讀入美元匯率歷史資料，並與上面計算出來的情緒分數依日期合併
USDTWD= pd.read_csv('USDTWD.csv')
df = pd.merge(USDTWD, trend,  how='outer',on=['Date'],sort=True)

#取到小數第三位，並將缺值補零
df=df.fillna(0)
df=df.round(3)
df
```

```python
#存成 final.csv
df.to_csv("final.csv",sep=',')
```
